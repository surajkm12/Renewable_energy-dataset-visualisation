## Questions & Answers

### Question 1.What is the distribution of the percentage of all renewables, excluding hydropower, for all countries in the dataset?

```python


df["Percentage of renewables excluding hydropower"] = df["non_hydropower"] / df["all_renewables"] * 100


plt.hist(df["Percentage of renewables excluding hydropower"], range= (0, 100))
plt.xlabel("Percentage of Renewables Excluding Hydropower")
plt.ylabel("Number of Countries")
plt.title("Distribution of Percentage of Renewables Excluding Hydropower")
plt.show()

print("Median percentage of renewables excluding hydropower:", df["Percentage of renewables excluding hydropower"].median())
print("Standard deviation of percentage of renewables excluding hydropower:", df["Percentage of renewables excluding hydropower"].std())
```

```python
df ['Percentage of renewables excluding hydropower'] = df ['Percentage of renewables excluding hydropower'].fillna (0)
df
```

### Question 2: What is the average percentage of electricity generation from all renewables across all countries in 2020?

```python

if 'Latest_Year' in df.columns:
    
    df['Latest_Year'] = pd.to_numeric(df['Latest_Year'].str.rstrip('%'), errors='coerce')

   
    if df['Latest_Year'].notnull().all():
       
        avg_percentage = df['Latest_Year'].mean()
        print(f"2. The average percentage of electricity generation from all renewables across all countries in 2020 was: {avg_percentage:.2f}%")
    else:
        print("2. Error: Conversion to numeric failed.")
else:
    print("2. 'all renewables 2020 or latest year' column not found in the DataFrame.")
```

### Question 3. How many countries had zero electricity generation from all renewables in 2020?

```python

if 'Latest_Year' in df.columns:
    zero_generation_countries = df[df['Latest_Year'] == 0]['Country'].tolist()
    print(f"3. {len(zero_generation_countries)} countries had zero electricity generation from all renewables in 2020.")
else:
    print("3. 'all renewables 2020 or latest year' column not found in the DataFrame.")
```

### Question 4. Visualize the top 5 countries with the highest electricity generation from all renewables in 2020.

```python

if 'electricity generation(Gwh)' in df.columns:
   
    top_countries_df = df.sort_values(by='electricity generation(Gwh)', ascending=False).head(5)
    
 
    plt.figure(figsize=(10, 6))
    plt.bar(top_countries_df['Country'], top_countries_df['electricity generation(Gwh)'], color='green')
    plt.xlabel('Country')
    plt.ylabel('Percentage of Electricity Generation from All Renewables (2020)')
    plt.title('Top 5 Countries with Highest Electricity Generation from All Renewables (2020)')
    plt.show()
else:
    print("4. 'all renewables 2020 or latest year' column not found in the DataFrame.")
```

### Questions 5. What is the total percentage of electricity generation from all renewables globally in 2020?

```python

if 'Latest_Year' in df.columns:
   
    total_percentage = df['Latest_Year'].sum()
    print(f"5. The total percentage of electricity generation from all renewables globally in 2020 was: {total_percentage}%")
else:
    print("5. 'all renewables 2020 or latest year' column not found in the DataFrame.")
```

### Question 6. Visualize the distribution of electricity generation from all renewables across all countries in 2020.

```python

if 'Latest_Year' in df.columns:
  
    plt.figure(figsize=(10, 6))
    plt.hist(df['Latest_Year'], bins=20, color='blue', edgecolor='black')
    plt.xlabel('Percentage of Electricity Generation from All Renewables (2020)')
    plt.ylabel('Number of Countries')
    plt.title('Distribution of Electricity Generation from All Renewables (2020) across Countries')
    plt.show()
else:
    print("6. 'all renewables 2020 or latest year' column not found in the DataFrame.")
```

### Question 7. countries with the highest total electricity generation:



```python

if 'electricity generation(Gwh)' in df.columns:
    
    max_generation_country = df.loc[df['electricity generation(Gwh)'].idxmax()]['Country']
    max_generation_value = df['electricity generation(Gwh)'].max()

    print(f"7. Which country has the highest total electricity generation in 2020 or the latest year? \nAnswer: {max_generation_country} with {max_generation_value} gigawatt-hours.")
else:
    print("7. 'total electricity generation, gigawatt-hours (gwh)' column not found in the DataFrame.")

```

### Question 10.What is the trend of renewable energy consumption in the top 3 countries with the highest consumption?

```python
top3_countries = df.groupby('Country')['all_renewables'].sum().nlargest(3).index
top3_countries_data = df[df['Country'].isin(top3_countries)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=top3_countries_data, x='Latest_Year', y='all_renewables', hue='Country', marker='o')
plt.title('Trend of Renewable Energy Consumption in Top 3 Countries')
plt.xlabel('Year')
plt.ylabel('Renewable Energy Consumption (Quadrillion Btu)')
plt.show()
```

### Question 12.Which countries have the highest and lowest percentages of all renewables, excluding hydropower?

```python

df_sorted = df.sort_values('non_hydropower', ascending=False)
highest_country = df_sorted.iloc[0]['Country']
highest_percentage = df_sorted.iloc[0]['non_hydropower']


lowest_country = df_sorted.iloc[-1]['Country']
lowest_percentage = df_sorted.iloc[-1]['non_hydropower']

print(f"The country with the highest percentage of renewables (excluding hydropower) is {highest_country} with {highest_percentage}%.")
print(f"The country with the lowest percentage of renewables (excluding hydropower) is {lowest_country} with {lowest_percentage}%.")
```

### Question 13.Which countries have a percentage of all renewables, excluding hydropower, less than 10%?

### Question 14.Which countries have a percentage of all renewables, excluding hydropower, between 70% and 80%?

### Question 15.Create a bar chart to compare the hydropower percentage of the top 10 countries with the highest hydropower percentage.

```python

sorted_df = df.sort_values(by='Latest_Year', ascending=False)


top_10_countries = sorted_df.head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['Country'], top_10_countries['Latest_Year'], color='blue')
plt.xlabel('Country')
plt.ylabel('Hydropower Percentage')
plt.title('Top 10 Countries with Highest Hydropower Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

 ### Question 16.  What is the correlation between the percentage of renewables in 2020 and the percentage difference between 2015 and 2020?

```python
df['Latest_Year'] = pd.to_numeric(df['Latest_Year'], errors='coerce')
df['difference(2015 vs 2020)'] = pd.to_numeric(df['difference(2015 vs 2020)'], errors='coerce')
correlation = df['Latest_Year'].corr(df['difference(2015 vs 2020)'])
print(f"The correlation between the percentage in 2020 and the percentage difference is: {correlation}")
```
